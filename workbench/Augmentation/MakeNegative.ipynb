```python
import os, fnmatch
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
from scipy.misc import imresize
import cv2
import random

%matplotlib inline

def find_files(directory, pattern): # from stackoverflow
    for root, dirs, files in os.walk(directory):
        for basename in files:
            if fnmatch.fnmatch(basename, pattern):
                filename = os.path.join(root, basename)
                yield filename
                

negative_source = "image_color_topic/"

def image_in_list(img):
    bb = 0
    hh = 0
    
    breite = 85
    hoehe = 256

    if len(img.shape) == 3:
        hh, bb, _ = img.shape
        
    b1 = random.randint(0, bb - breite)
    h1 = random.randint(0, hh - hoehe)
    
    neg = img[h1:h1+hoehe,b1:b1+breite,]
    return neg
        
def load_and_extract(directory, fn_handler, feature_list):
    shape = None        
    cnt = 0
    
    for filename in find_files(directory, '*.jpg'):
        cnt += 1
        img = plt.imread(filename)
        #img = img.astype(np.uint8)
        
        if len(img.shape) > 2 and img.shape[2] == 4:
            #converthe image from RGBA2RGB
            img = cv2.cvtColor(img, cv2.COLOR_RGBA2RGB)
        
        feature_list.append(fn_handler(img))
        
X_negative = []

load_and_extract(negative_source, image_in_list, X_negative)

for i in range(0, len(X_negative)):
    img_hsv=cv2.cvtColor(X_negative[i], cv2.COLOR_RGB2HSV)

    # lower mask (0-10)
    lower_red = np.array([0,50,50])
    upper_red = np.array([10,255,255])
    mask0 = cv2.inRange(img_hsv, lower_red, upper_red)

    # upper mask (170-180)
    lower_red = np.array([170,50,50])
    upper_red = np.array([180,255,255])
    mask1 = cv2.inRange(img_hsv, lower_red, upper_red)

    lower_green = np.array([46,50,50])
    upper_green = np.array([100,255,255])
    mask2 = cv2.inRange(img_hsv, lower_green, upper_green)
    
    # join my masks
    mask = mask0+mask1+mask2

    # set my output img to zero everywhere except my mask
    cnt = np.count_nonzero(mask)
    
    if cnt == 0:
        cv2.imwrite("negatives/{}.jpg".format(i), cv2.cvtColor(X_negative[i], cv2.COLOR_RGB2BGR))
    else:
        print("rot")
```
